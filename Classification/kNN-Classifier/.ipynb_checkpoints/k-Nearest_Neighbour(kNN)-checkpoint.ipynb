<h5>Model Building</h5>

Splitting the data before the Scaling operations as Outcome need not be scaled</blockquote>

```python
X = df_copy.drop(columns=['Outcome'],axis=1)
Y = df_copy['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

<blockquote>Scaling and Normalization Operations</blockquote>

```python
from sklearn.preprocessing import StandardScaler
# define standard scaler
scaler = StandardScaler()
df_toscale = X
df_toscale.shape
```

```python
#Performing Scaling Operations in order to get better results
scale = scaler.fit_transform(df_toscale.values)
df_scaled = pd.DataFrame(scale, index=df_toscale.index, columns=df_toscale.columns)
```

```python
df_scaled
```

```python
X = df_scaled
X.head()
```

```python
#Splitting the data in Training and Testing samples
#Importing Train-Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=42,stratify=Y)
```

```python
#Importing kNN Neighbour Classifier
from sklearn.neighbors import KNeighborsClassifier

neighbors = np.arange(1,30)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setting up a kNN Classifiers with k Neighbours
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #Fitting the Model
    knn.fit(X_train,Y_train)
    
    #Computing Accuracy on the Training Set
    train_accuracy[i] = knn.score(X_train, Y_train)
    
    #Compute the Accuracy on the Test Set
    test_accuracy[i] = knn.score(X_test, Y_test)
    test_accuracy[i]
    print("Accuracy for",i+1,"is",test_accuracy[i])
```

```python
#Generate Plot

plt.title('kNN Varying no: of Neighbours')
plt.plot(neighbors,test_accuracy,label = 'Testing Accuracy')
plt.plot(neighbors,train_accuracy,label = 'Training Accuracy')
plt.xticks(range(1,30))
plt.legend()
plt.xlabel('Number of Neighbours')
plt.ylabel('Accuracy')
plt.show()
```

```python
#Setup kNN Classifier with k neighbours
knn = KNeighborsClassifier(n_neighbors=14)
#Fit the Model
knn.fit(X_train,Y_train)
```

```python
#Calculating the Accuracy of the Classifier 
knn.score(X_test,Y_test)
```

<h5>Confusion Matrix</h5>

```python
#Importing the Confusion Matrix library
from sklearn.metrics import confusion_matrix
```

```python
#Getting the Predicted values using Classifier
Y_pred = knn.predict(X_test)
confusion_matrix(Y_test,Y_pred)
```

```python
#Using the Pandas's confusion matrix lib which gives a better representation
```

```python
pd.crosstab(Y_test,Y_pred, rownames=['True'],colnames=['Predicted'],margins=True)
```

Printing the True Negatives, False Positives, False Negatives, True Positives metrics

```python
tn, fp, fn, tp = confusion_matrix(Y_test,Y_pred).ravel()
print("True Negatives:",tn)
print("False Positives:",fp)
print("False Negatives:",fn)
print("True Positives:",tp)
```

```python
#Manually Calculating the Accuracy
accr = (tp+tn)/(tn+fp+fn+tp)
print("The accuracy of the model is",accr)
```

<h7>Precision</h7>

Precision tells us how many of the correctly predicted cases actually turned out to be positive.

```python
prec = tp/(tp+fp)
print("The Precision of the model is",prec)
```

<h7>Recall</h7>

Recall tells us how many of the actual positive cases we were able to predict correctly with our model.

```python
recall = tp/(tp+fn)
print("The Recall of the model is",recall)
```

<h7>F1 - Score</h7>

```python
from sklearn.metrics import f1_score
```

```python
f1_bin = f1_score(Y_test,Y_pred,average='binary')
f1_micro = f1_score(Y_test,Y_pred,average='micro')
f1_none = f1_score(Y_test,Y_pred,average=None)
print("\n","f1 Score Binary",f1_bin,"\n","f1 Score Micro:",f1_micro,"\n","f1 Score None",f1_none)
```

```python
Y_pred_proba = knn.predict_proba(X_test)[:,1]
Y_predict = knn.predict(X_test)
```

<h5>Hyperparameter Tuning</h5>

```python
#Importing the GridSearchCV package 
from sklearn.model_selection import GridSearchCV
```

```python
#Tuning the n_neighbours parameter as it is the one to be trained for a Classifier
param_grid = {'n_neighbors':np.arange(1,50)}
```

```python
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn,param_grid,cv=10)
knn_cv.fit(X,Y)
```
